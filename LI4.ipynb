**Média de gols de times como mandante de um jogo dos jogos entre 2003 e 2022**


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("campeonato-brasileiro-full.csv")
media = df['mandante_Placar'].mean()
print(f'A média de gols de times como mandante" é: {media}')
```

**CALCULO DA MEDIA DE GOLS POR TIME MANDANTE NO ANO DE 2021**

```python
import pandas as pd
import matplotlib.pyplot as plt

df['data'] = pd.to_datetime(df['data'])
partidas_2021 = df[df['data'].dt.year == 2021]
media_gols_2021 = partidas_2021['mandante_Placar'].mean()
print(f'A média de gols em 2021 foi de {media_gols_2021:.2f}')

```

**Grafico top 10 atletas com mais gols entre 2003 - 2022**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

contagem_gols = df['atleta'].value_counts().head(10)

rótulos = [f'{atleta} ({contagem})' for atleta, contagem in zip(contagem_gols.index, contagem_gols)]

sns.set(style="whitegrid")

fig, ax = plt.subplots(figsize=(10, 6))

cores = sns.color_palette("viridis", len(contagem_gols))
plt.barh(rótulos, contagem_gols, color=cores)
plt.title('Top 10 Atletas com Mais Gols', fontsize=16)
plt.xlabel('Contagem de Gols', fontsize=14)
plt.ylabel('Atleta', fontsize=14)
plt.gca().invert_yaxis()

ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

ax.grid(axis='x', linestyle='--', alpha=0.6)

sns.despine()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

contagem_gols_por_clube = df['clube'].value_counts().reset_index()
contagem_gols_por_clube.columns = ['Clube', 'Total de Gols']

contagem_gols_por_clube = contagem_gols_por_clube.sort_values(by='Total de Gols', ascending=False)

sns.set(style="whitegrid")

fig, ax = plt.subplots(figsize=(12, 6))

sns.barplot(x='Total de Gols', y='Clube', data=contagem_gols_por_clube, palette='viridis', ax=ax)

ax.set_title('Contagem de Gols por Clube', fontsize=16)
ax.set_xlabel('Total de Gols', fontsize=14)
ax.set_ylabel('Clube', fontsize=14)

ax.bar_width = 0.6

for i, v in enumerate(contagem_gols_por_clube['Total de Gols']):
    ax.text(v, i, str(v), ha='left', va='center', fontsize=12)

sns.despine(left=True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

gols_contra = df[df['tipo_de_gol'] == 'Gol Contra']

contagem_gols_contra_por_clube = gols_contra['clube'].value_counts().reset_index()
contagem_gols_contra_por_clube.columns = ['Clube', 'Total de Gols Contra']

contagem_gols_contra_por_clube = contagem_gols_contra_por_clube.sort_values(by='Total de Gols Contra', ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(contagem_gols_contra_por_clube['Clube'], contagem_gols_contra_por_clube['Total de Gols Contra'], color='indianred')
plt.title('Gols Contra por Clube', fontsize=16)
plt.xlabel('Clube', fontsize=14)
plt.ylabel('Total de Gols Contra', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)

for i, v in enumerate(contagem_gols_contra_por_clube['Total de Gols Contra']):
    plt.text(i, v, str(v), ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```
